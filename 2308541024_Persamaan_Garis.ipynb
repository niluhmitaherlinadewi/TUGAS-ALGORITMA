```python
#Jika ada dua persamaan garis maka berapa solusi yang akan terjadi
#Jika m1 == m2
#1. Memanggil numpy
import numpy as np
#2. Memanggil pyplot
import matplotlib.pyplot as plt
#3. Masukkan nilai dari dua persamaan y1 = (a,b),(c,d) dan y2 = (e,f),(g,h)
    #Persamaan garis pertama y1 = m1*(x-a)+b
p_24 = 1
q_24 = 2
r_24 = 2
s_24 = 3
m1_24=(s_24-q_24)/(r_24-p_24)
print("Gradien garis pertama adalah {}".format(m1_24))
print("Persamaan garis yang pertama adalah y={}x+{} ".format(m1_24,-m1_24*p_24+q_24))

    #Persamaan garis kedua y2 = m2*(x-e)+f
j_24 = 4
k_24 = 5
l_24 = 5
m_24 = 6
m2_24=(m_24-k_24)/(l_24-j_24)
print("Gradien garis  adalah kedua {}".format(m2_24))
print("Persamaan garis yang kedua adalah y={}x+{} ".format(m2_24,-m2_24*j_24+k_24))

#4. permisalan
if m1_24==m2_24:
      print("Gradiennya sama, sehingga bisa jadi persamaan garis ini memiliki banyak solusi atau tidak memiliki solusi")
else:
      #Memiliki titik potong
      tp_x_24 = (m1_24*p_24-q_24-m2_24*j_24+k_24)/(m1_24-m2_24)
      tp_y_24 = m2_24*(tp_x_24-j_24)+k_24
      print("Titik potongnya adalah ({},{})".format(tp_x_24,tp_y_24))
#5. Menentukan range dari x
      x_24=np.arange(tp_x_24-5,tp_x_24+5,0.01)
      plt.figure(figsize=(4,4),dpi=500)

#6. Menunjukkan hasil
      plt.plot(x_24,m1_24*(x_24-p_24)+q_24,color='blue',label='y={}x+{}'.format(m1_24,-m1_24*p_24+q_24))
      plt.plot(x_24,m2_24*(x_24-j_24)+k_24,color='green',label='y={}x+{}'.format(m2_24,-m2_24*j_24+k_24))
      plt.plot(tp_x_24,tp_y_24,color='red',marker='o',markersize=4)
      plt.legend(loc='best')
      plt.show()
```
