```python
berat_beras = float(input("Berapakah berat satu karung beras? "))
kg_atau_g = input("Beratnya dalam kg atau g? ")

if kg_atau_g.lower() == 'kg':
    berat_beras = berat_beras*1000
    print("Jika berat beras dari kg menjadi g adalah  " + str(berat_beras) + "g")
elif kg_atau_g.lower() == 'g':
    berat_beras = berat_beras/1000
    print(f"Jika berat beras dari g menjadi kg adalah {berat_beras} kg")
```

```python
buku = float(input("Berapakah buah buku yang dibeli Rina? "))
lusin_atau_gross = input("per lusin atau per gross? ")

if lusin_atau_gross.lower() == 'lusin':
    buku = buku/12
    print("Jumlah buku dalam satuan lusin " + str(buku) + " lusin")
elif lusin_atau_gross.lower() == 'gross':
    buku = buku/144
    print(f"Jumlah buku dalam satuan gross {buku} gross")
```
