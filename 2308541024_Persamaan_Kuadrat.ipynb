```python
# Project 1 (Menentukan akar-akar dari persamaan ax**2 + bx +  c = 0)
# Menentukan akar-akar bisa dengan faktorisasi atau rumus x1,x2
# Langkah-langkah menentukan akar-akar persamaan kuadrat adalah :
# 1. Memanggil numpy
import numpy as np
# 2. Memanggil pyplot
import matplotlib.pyplot as plt
# 3. Memanggil cmath
import cmath
# 4. Input nilai a,b dan c dalam persamaan ax**2 + bx +  c = 0
m_24 = (float(input("Nilai konstanta m_24 adalah ")))
i_24 = (float(input("Nilai konstanta i_24 adalah ")))
t_24 = (float(input("Nilai konstanta t_24 adalah ")))
if m_24 != 0:
  print("Nilai yang anda input bisa membentuk persamaan kuadrat")
else:
  print("Error, nilai yang anda input bukan persamaan kuadrat")
print("Persamaan Kuadratnya adalah ({}x**2+{}x+{})".format(m_24,i_24,t_24))
# 5. Input rumus diskriminan
D_24 = (i_24**2-4*m_24*t_24)
# 6. Input rumus x1,x2
import cmath
x1_24 = (-i_24 + cmath.sqrt(D_24))/(2*m_24)
x2_24 = (-i_24 - cmath.sqrt(D_24))/(2*m_24)
print("Akar-akarnya adalah ({:.0f},{:.0f})".format(x1_24,x2_24))
```

```python
# Project 3.1 Menentukan titik puncak pada Persamaan Kuadrat
# 1. Memanggil numpy
import numpy as np
# 2. Memanggil pyplot
import matplotlib.pyplot as plt
# 3. Memanggil cmath
import cmath

# 4. Input nilai a,b,c
m_24 = float(input("Nilai konstanta m_24 adalah: "))
i_24 = float(input("Nilai konstanta i_24 adalah: "))
t_24 = float(input("Nilai konstanta t_24 adalah: "))

# 5. Menghitung titik puncak dan sumbu simetri
def menghitung_titik_puncak_dan_sumbu_simetri(m_24, i_24, t_24):
    # Sumbu simetri (x) adalah -b / (2a)
    sumbu_simetri_24 = -i_24 / (2 * m_24)
    # Titik puncak (x, y)
    # x dari titik puncak adalah sumbu simetri
    # y dari titik puncak adalah nilai fungsi kuadrat di x = sumbu_simetri
    titik_puncak_x_24 = sumbu_simetri_24
    titik_puncak_y_24 = m_24 * (titik_puncak_x_24 ** 2) + i_24 * titik_puncak_x_24 + t_24

    return (titik_puncak_x_24, titik_puncak_y_24), sumbu_simetri_24

# 6. Menentukan grafik persamaan kuadrat
def grafik_persamaan_kuadrat(m_24, i_24, t_24):
    titik_puncak_24, sumbu_simetri_24 = menghitung_titik_puncak_dan_sumbu_simetri(m_24, i_24, t_24)

    # Menentukan rentang x untuk grafik
    x_24 = np.linspace(titik_puncak_24[0] - 10, titik_puncak_24[0] + 10, 400)
    y_24 = m_24 * x_24**2 + i_24 * x_24 + t_24

    # Menginput Grafik persamaan kuadrat
    plt.figure(figsize=(10, 6))
    plt.plot(x_24, y_24, label=f'{m_24}x^2 + {i_24}x + {t_24}')

    # Input Titik puncak
    # Titik puncak berwarna hijau
    plt.plot(titik_puncak_24[0], titik_puncak_24[1], 'go')
    plt.annotate(f'Titik puncak\n({titik_puncak_24[0]:.2f}, {titik_puncak_24[1]:.2f})',
                 xy=titik_puncak_24,
                 xytext=(titik_puncak_24[0] + 1, titik_puncak_24[1] + 10),
                 arrowprops=dict(facecolor='black', shrink=0.05))
    plt.axvline(x=sumbu_simetri_24, color='gray', linestyle='--', label='Sumbu simetri')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Titik Puncak dari persamaan Kuadrat')
    plt.axvline(x=0, color='black')
    plt.axhline(y=0, color='black')
    plt.legend()
    plt.grid(True)
    plt.show()

# 7. Persamaan_kuadrat dengan nilai a, b, dan c yang diberikan
grafik_persamaan_kuadrat(m_24, i_24, t_24)
```
